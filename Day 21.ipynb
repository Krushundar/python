```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Load the dataset
data = pd.read_csv("Civil_Engineering_Regression_Dataset.csv")  # Replace 'your_dataset.csv' with your dataset's path

# Step 2: Display the first few rows of the dataset
print("First few rows of the dataset:")
print(data.head())

# Step 3: Identify independent and dependent variables
# Assuming 'Construction Cost' is the dependent variable and other columns are independent variables
dependent_var = 'Construction Cost'  # Replace with your dependent variable column
independent_vars = data.columns[data.columns != dependent_var]  # Independent variables are all other columns

print("\nIndependent Variables:", independent_vars)
print("\nDependent Variable:", dependent_var)

# Step 4: Check for missing values
missing_values = data.isnull().sum()

print("\nMissing values in each column:")
print(missing_values)

# Step 5: Handle missing values
# For numerical columns, we will fill missing values with the median
data.fillna(data.median(), inplace=True)

# If you have categorical columns, you can fill missing values with the mode (optional)
# data['Category Column'] = data['Category Column'].fillna(data['Category Column'].mode()[0])

# Verify that missing values have been handled
print("\nMissing values after handling:")
print(data.isnull().sum())

# Step 6: Generate summary statistics for numerical variables
summary_stats = data.describe()

print("\nSummary Statistics (Mean, Median, Std, etc.):")
print(summary_stats)

# Step 7: Create a correlation heatmap
# Calculate the correlation matrix
correlation_matrix = data.corr()

# Plot the heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5)

# Display the heatmap
plt.title('Correlation Heatmap')
plt.show()

# Step 8: Interpretation of correlation between Construction Cost and other factors
construction_cost_corr = correlation_matrix[dependent_var]

print(f"\nCorrelation between Construction Cost and other variables:\n{construction_cost_corr}")
```
