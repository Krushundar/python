```python
import statsmodels.formula.api as smf
model = smf.ols("sunday~daily",data = data1).fit()
```

```python
model.summary()
```

#### Obseravtion
- The predicted equation is y_hat = Beta_0 + Beta_1(x)
- beta_0 = 13.8356 , beta_1 = 1.33887(x)      

- The probabilty(p-value) for intercept(beta_0) is 0.707 > 0,05
- Therefore the intercept coefficent may not be that much significant in prediction
- However the p-valuee for "daily" (beta_1) is 0.00 < 0.05
- Therefore the beta_1 coefficient is highly significant and is contributint to prediction.

```python
import statsmodels.formula.api as smf
model = smf.ols("sunday~daily",data = data1).fit()
```

```python
model.params
```

```python
print(f'model t-values:\n{model.tvalues}\n----------\nmodel p-values: \n{model.pvalues}')
```

```python
(model.rsquared,model.rsquared_adj)
```

```python
newdata=pd.Series([200,300,1500])
```

```python
data_pred=pd.DataFrame(newdata,columns=['daily'])
data_pred
```

```python
model.predict(data_pred)
```

```python
pred = model.predict(data1["daily"])
pred
```

```python
data1["Y_hat"] = pred
data1
```

```python
data1["residuals"] = data1["sunday"]-data1["Y_hat"]
data1
```

```python
mse = np.mean((data1["daily"]-data1["Y_hat"])**2)
rmse = np.sqrt(mse)
print("MSE: ",mse)
print("RMSE: ",rmse)
```

#### Checking the model residuals scatters plot

```python
plt.scatter(data1["Y_hat"], data1["residuals"])
```

#### Observation
- There appears to be no trend and the residuals are randomly placed around the zero error line
- Hence the assumpmtion of homoscedasticty is satisfied (constant variance in residuals)

```python
import statsmodels.api as sm
sm.qqplot(data1["residuals"], line = '45', fit = True)
plt.show()
```
